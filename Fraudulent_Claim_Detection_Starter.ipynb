## **Fraudulent Claim Detection**

## Problem Statement
Global Insure, a leading insurance company, processes thousands of claims annually. However, a significant percentage of these claims turn out to be fraudulent, resulting in considerable financial losses. The company’s current process for identifying fraudulent claims involves manual inspections, which is time-consuming and inefficient. Fraudulent claims are often detected too late in the process, after the company has already paid out significant amounts. Global Insure wants to improve its fraud detection process using data-driven insights to classify claims as fraudulent or legitimate early in the approval process. This would minimise financial losses and optimise the overall claims handling process.

## Business Objective
Global Insure wants to build a model to classify insurance claims as either fraudulent or legitimate based on historical claim details and customer profiles. By using features like claim amounts, customer profiles and claim types, the company aims to predict which claims are likely to be fraudulent before they are approved.

#### **2.2.3** Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values <font color="red">[1 Mark]</font>

```python
# Identify and drop rows where features have illogical or invalid values, such as negative values for features that should only have positive values

# List of columns that should only have positive values
# (assuming these are the columns based on the dataset description)

positive_cols = [
    'months_as_customer', 'age', 'policy_deductable', 'policy_annual_premium',
    'umbrella_limit', 'capital-gains', 'capital-loss', 'total_claim_amount',
    'injury_claim', 'property_claim', 'vehicle_claim',
    'number_of_vehicles_involved', 'bodily_injuries', 'witnesses'
]


negative_mask = df[positive_cols] < 0
negative_counts = negative_mask.sum()

print("\nNegative value count per column:")
print(negative_counts[negative_counts > 0])

if negative_mask.any(axis=1).any():
    print("\nRows with remaining negative values:")
    print(df.loc[negative_mask.any(axis=1), positive_cols])
else:
    print("\nNo remaining negative values in specified columns.")

# Replace negative values with 0 or median based on the column's context

cols_to_handle = {
    'capital-gains': 'replace_with_0',
    'umbrella_limit': 'replace_with_0',
    'policy_deductable': 'replace_with_median',
    'policy_annual_premium': 'replace_with_median',
    'injury_claim': 'replace_with_median',
    'property_claim': 'replace_with_median',
    'vehicle_claim': 'replace_with_median',
    'total_claim_amount': 'replace_with_median',
    'number_of_vehicles_involved': 'replace_with_0',
    'bodily_injuries': 'replace_with_0',
    'witnesses': 'replace_with_0'
}

# Clean negative values based on rules
for col, method in cols_to_handle.items():
    if col not in df.columns:
        print(f"Column '{col}' not found in DataFrame. Skipping.")
        continue

    if (df[col] < 0).any():
        if method == 'replace_with_0':
            df[col] = df[col].clip(lower=0)
        elif method == 'replace_with_median':
            median_val = df[col].median()
            df[col] = np.where(df[col] < 0, median_val, df[col])
        else:
            print(f"Unknown method '{method}' for column '{col}'. Skipping.")

# Optional summary check
for col in cols_to_handle:
    if col in df.columns:
        print(f"\nSummary for '{col}':")
        print(df[col].describe())

# Detect placeholder strings in object columns
placeholders = ['?', 'NA', 'N/A', 'null', 'None', 'none', '', ' ', '-', '--']
placeholder_counts = {}

for col in df.select_dtypes(include='object').columns:
    count = df[col].isin(placeholders).sum()
    if count > 0:
        placeholder_counts[col] = count

# Display placeholder summary
if placeholder_counts:
    print("\nColumns with placeholder values:")
    for col, count in placeholder_counts.items():
        print(f"{col}: {count} entries")
else:
    print("\nNo placeholder values found in object columns.")

# replacing placeholders with 'unknown' 
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].replace(placeholders, 'unknown')
```

Use the **RandomOverSampler** technique to balance the data and handle class imbalance. This method increases the number of samples in the minority class by randomly duplicating them, creating synthetic data points with similar characteristics. ### **6.3 Handle redundant columns** <font color="red">[3 marks]</font>

Analyse the data to identify features that may be redundant or contribute minimal information toward predicting the target variable and drop them.

- You can consider features that exhibit high correlation with other variables, which you may have observed during the EDA phase.
- Features that don't strongly influence the prediction, which you may have observed during the EDA phase.
- Categorical columns with low value counts for some levels can be remapped to reduce number of unique levels, and features with very high counts for just one level may be removed, as they resemble unique identifier columns and do not provide substantial predictive value.
- Additionally, eliminate any columns from which the necessary features have already been extracted in the preceding step.

```python
# Drop redundant columns from training and validation data



corr_matrix = X_train_resampled.select_dtypes(include=np.number).corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_cols = [column for column in upper.columns if any(upper[column] > 0.9)]

print("Highly correlated columns to drop:", high_corr_cols)

X_train_resampled.drop(columns=high_corr_cols, inplace=True, errors='ignore')
X_val.drop(columns=high_corr_cols, inplace=True, errors='ignore')

# Drop low information categorical features (where one category dominates > 95%)
low_info_cols = []
for col in X_train_resampled.select_dtypes(include='object').columns:
    if col in X_train_resampled:
        top_freq = X_train_resampled[col].value_counts(normalize=True).values[0]
        if top_freq > 0.95:
            low_info_cols.append(col)

print("Low information categorical columns to drop:", low_info_cols)

X_train_resampled.drop(columns=low_info_cols, inplace=True, errors='ignore')
X_val.drop(columns=low_info_cols, inplace=True, errors='ignore')


print("Remaining NaNs in train set:", X_train_resampled.isnull().sum().sum())
print("Remaining NaNs in val set:", X_val.isnull().sum().sum())
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information

print("Categorical columns in training data: ", X_train_resampled.select_dtypes(include='object').columns)
categorical_cols = X_train_resampled.select_dtypes(include='object').columns

for col in categorical_cols:
    if col in X_train_resampled.columns:
        value_counts = X_train_resampled[col].value_counts()
        low_freq = value_counts[value_counts < 10].index

        # Replace in train
        X_train_resampled[col] = X_train_resampled[col].replace(low_freq, 'Other')

        # Replace in val: map to 'Other' if not in valid categories
        valid_categories = value_counts.index.difference(low_freq)
        X_val[col] = X_val[col].apply(lambda x: x if x in valid_categories else 'Other')
```
